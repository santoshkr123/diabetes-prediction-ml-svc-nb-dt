```python
import numpy as np
import pandas  as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

```python
data=pd.read_csv('diabetes.csv')
```

```python
data.head()
```

```python
data.describe()
```

```python
data.isnull().sum()
```

```python
fig,ax=plt.subplots(figsize=(15,12))
sns.boxplot(data=data,ax=ax,fliersize=3,width=0.3)
```

```python
data.describe()
```

```python
data['BMI']=data['BMI'].replace(0,data['BMI'].mean())
data['BloodPressure']=data['BloodPressure'].replace(0,data['BloodPressure'].mean())
data['Glucose']=data['Glucose'].replace(0,data['Glucose'].mean())
data['SkinThickness']=data['SkinThickness'].replace(0,data['SkinThickness'].mean())
data['Insulin']=data['Insulin'].replace(0,data['Insulin'].median())

```

```python
data.describe()
```

```python
data.head()
```

```python
X=data.drop(columns=['Outcome'])
y=data['Outcome']
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=42)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler=StandardScaler()
```

```python
X_train_scaled=scaler.fit_transform(X_train)
X_test_scaled=scaler.transform(X_test)
```

```python
X_train_scaled
```

```python
X_test_scaled
```

# model building using DecisionTree

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
parameter={
    'criterion':['gini', 'entropy', 'log_loss'],
    'splitter':['best', 'random'],
    'max_depth':[1,2,3,4,5],
    'max_features':['sqrt','log2']
}
```

```python
classifier=DecisionTreeClassifier()
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
clf=GridSearchCV(classifier,param_grid=parameter,cv=3,scoring='accuracy',verbose=3)
```

```python
clf.fit(X_train_scaled,y_train)
```

```python
clf.best_params_
```

```python
classifier=DecisionTreeClassifier(criterion= 'entropy',
 max_depth= 2,
 max_features= 'log2',
 splitter='best')
```

```python
classifier.fit(X_train,y_train)
```

```python
y_pred=classifier.predict(X_test)
```

```python
y_pred
```

```python
y_test
```

```python
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
```

```python
print(accuracy_score(y_pred,y_test))
print(classification_report(y_pred,y_test))
print(confusion_matrix(y_pred,y_test))
```

# model building SVC

```python
from sklearn.svm import SVC
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
parameter={
    'C':[0.1,1,10],
    'gamma':[0.1,0.001,0.0001],
    'kernel':['linear','rbf','polynomial']
}
```

```python
grid=GridSearchCV(SVC(),param_grid=parameter,refit=True,cv=3,verbose=3,scoring='accuracy')
```

```python
grid.fit(X_train,y_train)
```

```python
grid.best_params_
```

```python
svc_clf=SVC(C= 1, gamma= 0.1, kernel= 'linear')
```

```python
svc_clf.fit(X_train_scaled,y_train)
```

```python
y_pred_svc=svc_clf.predict(X_test_scaled)
```

```python
y_pred_svc
```

```python
y_test
```

```python
from sklearn.metrics import classification_report,accuracy_score,confusion_matrix
```

```python
print(classification_report(y_pred_svc,y_test))
print(accuracy_score(y_pred_svc,y_test))
print(confusion_matrix(y_pred_svc,y_test))
```
